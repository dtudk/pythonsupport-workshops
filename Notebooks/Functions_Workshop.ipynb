#### Loops
A loop is a programming structure that allows a set of instructions to be executed repeatedly. It continues to run as long as a specified condition is true, making it useful for tasks that require repetition, such as processing items in a list or performing repetitive calculations. Common types of loops include for loops and while loops.

### Exercise 1: Multiplication

Imagine that we want to tell a computer how to multiply two numbers. We need to somehow provide with a concrete set of instructions. This means that we need to define:

1. How many inputs the function takes and what to call them 

2. What to do with the inputs

3. what to output

We will start off by doing this on a whiteboard for a procedure we all know, multiplication. The idea here is to try and break it down and see how we actually tell the computer what to do. 

### Exercise 2: Functions Exercise : multiplication


We want to implement a function, that takes three numbers $x$ and $y, multiplies them and saves the result. More concretely, we want to implement the function:

$f: (x,y) \rightarrow x \cdot y$

First, we need to begin the function by giving it a name such as "mult" or "multiply". Second, we need to name the inputs that it takes. This could be "x", "y" or anything. Lastly, we need to tell the function what to do. We can either, be explicit by letting the function calculate $x \cdot y $, save it and then return it, or we can just return $x \cdot y $ immediately. We want to break it down completely by viewing mulitplication of $x$ and $y$ as adding $x$ to itself $y$ times. Try and implement this below with a function that only makes additions.

```python
def multiply(x,y):
    product = 0
    for i in range(y):
        product+=x
    return product
```

```python
print(multiply(2,3))
```

### Exercise 3: Whiteboard power function

Now we want to see how we can make some procedures easier by recycling instruction we've already made. For this exercise, we will try to implement a power function on the whiteboard. The key idea here, is that raising a number to a power, can be broken down into multiplying that number by itself a bunch of times. Therefore we will implement the power function using the multiply function. 

### Exercise 4: Power function

Functions can (and often do) call other functions. As an example, let's make a function that uses our multiply function to raise a number to a power. Concretely we want to create the following function:


$f : (x,n) \rightarrow x^n$

Here, we can't just return the result immediately

```python
def power(x,n):
    result = 1
    for i in range(n):
        result=multiply(result,x)
    return result
```

```python
print(power(2,3))
```
