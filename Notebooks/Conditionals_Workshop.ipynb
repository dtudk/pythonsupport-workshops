#### Conditionals

Conditionals are a way of telling the computer what to do if something is true or false. This is done with if-statements. You can think of an if-statement as a question: if the answer is “yes” the condition is true and the computer runs the specified command. If the answer is “no” it skips to the next command. 
This is a good tool if you want to generalize your code to handle different cases or exceptions. The goal of your program may for example depend on what the user asks for, or how much input is received. 

### Exercise 1: Dice Game

Imagine that two people, player 1 and player 2 play a game of dice. The winner of the game is the person who rolls the highest number. If they the same number it's a tie and they roll again. We want to visualize this as a bunch of scenarios using the velcro cards. The trick here is the break it down into every possible scenario, and telling the computer what to do in each scenario. For example, one scenario is player 1 rolling higher than player 2. We want to make every scenario as a velcro card



### Exercise 2: Conditionals in python

Now we will code the dice game in python

```python
player1_roll = 5
player2_roll = 2

if player1_roll > player2_roll:
    print("Player 1 Wins!")
elif player1_roll < player2_roll:
    print("player 2 wins!")
else:
    print("Rematch")
```

Note that just as with the velcro cards, the order of the if statements does not really matter. We could just as well have gotten the same outputs by doing it in a different order.

```python
player1_roll = 5
player2_roll = 2

if player1_roll < player2_roll:
    print("Player 1 Wins!")
elif player1_roll > player2_roll:
    print("player 2 wins!")
else:
    print("Rematch")
```

### Exercise 4: Programming snakes and ladders

Now we want to transfer what we just did with the velcro cards onto the computer. We have helped a bit by already defining a list of the different positions. Your job is to assume that you are given a current position and a throw, and then make the computer print, which square you should move to

```python
positions = [
    "1",
    "2",
    "3",
    "ladder",
    "5",
    "6",
    "7",
    "snake",
    "9",
    "ladder",
    "11",
    "12",
    "13",
    "14",
    "15",
    "16",
    "17",
    "18",
    "ladder",
    "20",
    "21",
    "22",
    "23",
    "snake",
    "25",
]
```

```python
current_position = 3
roll = 5

if positions[current_position + roll - 1] == "snake":
    print("Go to square number:")
    print(current_position + roll - 6)
elif positions[current_position + roll - 1] == "ladder":
    print("Go to square number:")
    print(current_position + roll + 3)
else:
    print("Go to square number:")
    print(current_position + roll)
```
