#### Conditionals

Conditinals are a way of checking if something is true or not. This is done through what is called an if-statement. You can think of an if-statement as a question: if the answer is “yes” the condition is true and the computer runs the specified command. If the answer is “no” it skips to the next command. 
This is a good tool if you want to generalize your code to handle different cases or exceptions. The goal of your program may for example depend on what the user asks for, or how much input is received. 

### Exercise 1: Dice Game

Imagine that you and I play a gameof dice. The winner of the game is the person who rolls the highest number. If we roll the same number it's a tie and we roll again. 

Now how do we program this so a computer would be able to determine the outcome of the game? First we would need our two rolls as variables. We define that as Your_Roll and My_Roll which are two variables we can change the value of for each new game. 

Next thing we check is who's roll is highest. This can be done in any order but here we check if yours is the highest. This is done using if followed by our condition, which is if your roll is higher than mine ending with a ":". When pressing enter the next line will be slightly inwards on the page. The computer knows to run this line, if the condition above is true. If this condition is true it means that your roll is the highest and that you win. Therefore if this is true we will print "Your Win!". 

If your roll is not the highest we need to check if my roll is the highest. This is again done by using a conditional, this time we do it using "elif". This stands for else if and is called else if in other programming languages than Python. The else statement, which we will use later, executes something if all above if- and elif-statements are false. The elif statement checks a condition if the above if statement is false. The else statement does not check a condition before executing a command, whereas the elif statement only executes a command if the given condition is true. Thus the elif statement works like a normal if, except there need to be a prior if statement for it to work. If the prior if statement is true both elif- and else-statements will be ignored. If you in stead of an elif uses a second if-statement, the two conditionals will not be linked. This means that in some cases you can get multiple responses to something you only wanted one. 



```python
Your_Roll = 5
My_Roll = 2

if Your_Roll > My_Roll:
    print("You Win!")
elif Your_Roll < My_Roll:
    print("I Win!")
else:
    print("Rematch")
```
