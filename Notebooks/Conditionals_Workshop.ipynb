#### Conditionals

Conditinals are a way of checking if something is true or not. This is done through what is called an if-statement. You can think of an if-statement as a question: if the answer is “yes” the condition is true and the computer runs the specified command. If the answer is “no” it skips to the next command. 
This is a good tool if you want to generalize your code to handle different cases or exceptions. The goal of your program may for example depend on what the user asks for, or how much input is received. 

### Exercise 1: Dice Game

Imagine that you and I play a gameof dice. The winner of the game is the person who rolls the highest number. If we roll the same number it's a tie and we roll again. 



```python
Your_Roll = 5
My_Roll = 2

if Your_Roll > My_Roll:
    print("You Win!")
elif Your_Roll < My_Roll:
    print("I Win!")
else:
    print("Rematch")
```
