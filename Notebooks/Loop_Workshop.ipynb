#### Loops

A loop is a programming structure that allows a set of instructions to be executed repeatedly. It continues to run as long as a specified condition is true, making it useful for tasks that require repetition, such as processing items in a list or performing repetitive calculations. Common types of loops include for loops and while loops.

### Exercise 1: Jenga

Imagine that you and I are playing a game of Jenga, or in Danish, Klodsmajor. Let's say that you start. You take out a brick from the tower and check to see if the tower falls over. If not, it is my turn and I take out a brick from the tower. This continues until the tower falls over, at which point the game is over and the person who took out the last brick has lost. 

This can be described as what we in programming call a while loop. In a while loop the program continuously checks if a condition is true, and if so it executes a command repeatedly. 

Lets look at Jenga again. When it is someones turn they first look to see if the tower is still standing before taking a brick out. Thus our condition is: "Is the tower still standing?". If this condition is True then we can execute our command which is to take a brick from the tower. This continues until our condition is False, aka the tower has fallen over, and the game is over.

### Exercise 2: While Loop Computer Exercises

Now I am going to ask you something you have probably done many many times before in your life. I want you to count to 5. You probably start thinking 1,2,3 and so on in your head but what would a computer do? 

It would start from 0 and then add 1 until it hits 5. How can we do that as a while loop? First we would need to setup a condition to check. Here the first thought would be: "Is the count at 5?". But while loops will only execute a command when true, so when this condition is true we are already at our goal. Therefore the computer will not add to our counter unless the counter is already at 5. So how do we define our condition so it is false when the count hits 5?

There are two ways to do this. For the first example we will use the condition: "Is the count less than 5?". Here when we add one every time it will stop as soon as the counter is not less than 5, which means that it has hit 5. 

For this to work on a computer the first thing we need is a variable with a starting value. In this example we call the variable "Count" and set it equal to 0 as this is our starting point. 

The next thing we need to do is to initialize our while loop. In python we do that by, on a new line, writing "while" with a space after, followed by our condition using mathematical operators. In this example our condition "Is the count less than 5?" can be written as "Count < 5". After this we need to write a colon ":" and press enter to write our exercutions.

```python
# Removed before release
Count = 0

while Count < 5:
    Count = Count + 1
    print(Count)
```

### Exercise 3: Building with centicubes

**Row of 5 centicubes**

Now lets look at for loops. Imagine that you have a bunch of Centicubes and that you want to build a row. Lets say you want to build it with a single stack, 5 long. As before, we can think of it as a while loop, where we add one cube at a time until we hit 5 and the loop stops. We could also use a for loop instead. A for loop runs the code a pre defined amount of times. This means that if we want to build a row of five centicubes, we would start with 0, and then define a for loop to add a centicube 5 times. This can be written in python as:

```python
centicubes_needed = 0

for i in [0,1,2,3,4]:
    centicubes_needed += 1

print(centicubes_needed)
```

In python we usually use a smart function, ``range``. This function automatically goes through numbers 0 through the given number minus 1, i.e. ``range(5)`` gives the numbers ``0,1,2,3,4``. We use this instead of giving all loop indices:

```python
centicubes_needed = 0

for i in range(5):
    centicubes_needed += 1

print(centicubes_needed)
```

**5x5 square of centicubes**

Lets say that you now want to build the row in to a square. With centicubes the easiest way to do so is to build a bunch of rows and then put them together into a square. If we want a 5x5 square, we would start by building one row of 5, then another row of five and add it on top of the first wor. Then we would build a third row of five and add on top of the others and so on until we have 5 rows together as a $5\times5$ square. 

This is called a nested loop, meaning we have a loop inside of another loop. The inner loop builds the rows, and the outer loop controlles how many rows to build. 

This can be shown in python. Here instead of building the rows, it counts the centivubes but it works the same way.

```python
centicubes_needed = 0

for i in range(5):
    for j in range(5):
        centicubes_needed += 1

print(centicubes_needed)
```

**5x5x5 cube of centicubes**

Now that we hve built a square of 5x5 we want to make it three dimentional, so a cube. This is easiest done by building a new 5x5 square and adding it on top of the first square as a new layer. If we want a 5x5x5 we would need 5 layers of 5 squares.

To this we add an other loop to our two nested loops from before. This loop would be added as the outer most loop, controlling how many squares are build. To sum up, the innner loop is building a row, the middle loop is building a square from the rows, and the outer loop is building a cube from the squares. 

This can again be shown in python by counting the number of centicubes needed.

```python
centicubes_needed = 0

for i in range(5):
    for j in range(5):
            for k in range(5):
                centicubes_needed += 1

print(centicubes_needed)
```

**for loops with lists**

Another thing one can do using loops, is going through lists. Now imagine that you have a list of numbers and you want to add 1 to all of them. A for loop for this kind of task is beneficial. In the previous loops we used "for i in range(5)", but what if our list is longer than 5? Then we would never visit the last entries of the list. Instead we can take the length of the list as our range. Lets say that our list is called A and we want to find the length of it, we can use the len() function.

```python
A = [2,4,7,2,5,4,6,9,2]

len(A)
```

Try to add or remove entires and recalculate the lenght to see how it changes. 

So how do we use this in a loop? Instead of the "5" from before we simply put "len(A)". 

Now the "i" in the loop serves as a placeholder for what number loop we are running. This means that the first time the loop runs i is equal to 0. The next loop it is equal to 1 and so on, until the loop has run the given amount of times. This we can show by simply printing i in every loop:

```python
for i in range(len(A)):
    print(i)
```

This we can use when running through a list. Here we can go through the list index' one by one, using the value for i in stead of the index in A. Since the loop is runnning according to how many items are in A, the last value for i will match with the last index for A. Therefore if we want to add 1 to all numbers in A it would look something like this:

```python
A = [2,4,7,2,5,4,6,9,2]

for i in range(len(A)):
    A[i] += 1

print(A)
```
