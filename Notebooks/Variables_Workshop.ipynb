#### Variables
A variable is a way to save something in the computer, for it to be used again later. If we need the computer to remember a number, a word, or something more complex as a list, we save it in a variable.


### Exercise 1: Variables Intro

First we need to create a variable. This is done by unfolding a box and giving the variable a value by putting an item in the box. Lets make the variable A and set it equal to Banana. This is done by unfolding box A and putting a Banana in the box. Now lets create the variable B and give it the value Apple. 

Now set A equal to B, what happens? 

Well B stays the same, and A is equal to the value of B, which is Apple, thus A equals Apple. 

Now lets start again with A equals to Banana and B equals to Apple. How can we swap the values of the two variables without loosing the value of one in the process?

This is done by creating an extra in between variable. In this example we call this variable C. First we set C equal to A, to "save" the value of A. Then we set A equal to B, and at last we set B equal to C. 

### Exercise 2: Variable Change Python

Now lets see how this works in python. Lets first create two variable A and B, and set them equal to Banana and Apple. Remember when setting a variale equal to text or strings as it is called in coding language, to put "" around the text. 

You can see what your variables are equal to by writing print(A). In this example it is the variable A we are showing.

```python
A = "Banana"
B = "Apple"
print(A, B)
```

Now lets see what happens if we set A. equal to B

```python
# Removed before release
A = B
print(A, B)
```

Here we have lost the value of A, which cannot be restored.

Now lets do a variable change without loosing the values. Start with A equals to Banana and B equals to Apple. Then try to change the values of the two variables without loosing the value of one of them. 

Hint: You just did it in exercise 1

```python
# Remove before release
A = "Banana"
B = "Apple"
C = A
A = B
B = C

print(A, B)
```

### Exercise 3: Variable Operations

Now lets look at how variables can be used with multiple values, which we call a list. Here we can for example have A equal to Banana, Carrot and Watermelon. This is done by putting them in a list, and indexing them. 

Put two walls in box A. In the first room put a Banana, in the next room a Carrot and in the last room a Watermelon. To keep track of which value is where, they get indexed. The first value is at index 0, then the next at index 1, and so on. So looking at our example Banana would be at index 0, which we can write as A[0] = Banana. 

Now, what do you think would happen if we say that A[0] = Orange ?

Exactly as earlier, Orange would go in and replace Banana at A[0], but this time the rest of the values in the list are still the same. 

What would happen if we set A[2] = A[1] ?

Well A[1] is Carrot which means that both A[1] and A[2] are equal to Carrot, and A would be equal to [Orange, Carrot, Carrot].

### Exercise 4: Variable Operations Python

Lets look at how variables as lists work within programing. First make a list called A and set it equal to Banana, Carrot and Watermelon and print the vairable. Remember when writing a string, to put "" around it!

```python
# Removed before release
A = ["Banana", "Carrot", "Watermelon"]
print(A)
```

Lets see what happens if we set A[0] equal to Orange.

```python
# Removed before release
A[0] = "Orange"
print(A)
```

Now lets set A[2] equal to A[1].

```python
# Removed before release
A[2] = A[1]
print(A)
```

Lets dig a little bit deeper in how to work with lists. First create an empty list called A. This is done by writing the variable name equal to []. Print the list to see how it looks after each step.

```python
# Removed before release
A = []
print(A)
```

Now lets add some values to A. In stead of putting the values at specific spots like before, lets just add them one at a time. In python this can be done using the append function. I will show you how to use it. Let us first add Watermelon.


Append is ONLY a Python function and will not work in other programing languages!

```python
# Removed before release
A.append("Watermelon")
print(A)
```

Now try to add the following to the list:

25 , Orange , Milk

```python
# Removed before release
A.append("Orange")
A.append(25)
A.append("Milk")
print(A)
```

Now what happens if you set A[1] equal to Carrot?

```python
# Removed before release
A[1] = "Carrot"
print(A)
```
