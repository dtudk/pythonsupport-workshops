#### Loops
A loop is a programming structure that allows a set of instructions to be executed repeatedly. It continues to run as long as a specified condition is true, making it useful for tasks that require repetition, such as processing items in a list or performing repetitive calculations. Common types of loops include for loops and while loops.

### Exercise 1: Jenga

Imagine that you and I are playing a game of Jenga, or in Danish, Klodsmajor. Lets say that you start. You take out a brick from the tower and cehcks to see if the tower falls over. If not it is my turn and I take out a brick from the tower. This continues until the tower falls over at which point the game is over and the person who took out the last brick has lost. 

This can be described as what we in programming call a while loop. In a while loop the program checks if a condition is true, and if so it executes a command. 

Lets look at Jenga again. When it is someones turn they first look to see if the tower is still standing before taking a brick out. Thus our condition is: "Is the tower still standing?". If this condition is True then we can execute our command which is to take a brick from the tower. This continues until our condition is False, aka the tower has fallen over, and the gae is over.

### Exercise 2: While Loop Computer Exercises

Now I am going to ask you something you have probably done many many times before in your life. I want you to count to 5. You probably start thinking 1,2,3 and so on in your head but what would a computer do? 

It would start from 0 and then add 1 until it hit 5. How can we do that as a while loop? First we would need a condition. Here the first thought would be: "Is the count at 5?". But while loops will only exercute a command when true, so when this condition is true we are already at our goal. Therefore the computer will not add to our counter unless the counter is already at 5. So how do we define our condition so it is false when the count hits 5?

There are two ways to do this. For the first example we will use the condition: "Is the count less than 5?". Here when we add one every time it will stop as soon as the counter is not less than 5, which means that it has hit 5. 

For this to work on a computer the first thing we need is a variable with a starting value. In this example we call the variable "Count" and set it equal to 0 as this is our starting point. 

The next thing we need to do is to initialize our while loop. In python we do that by, on a new line, writing "while" with a space after, followed by our condition using mathematical operators. In this example our condition "Is the count less than 5?" can be written as "Count < 5". After this we need to write a colon (:) and press enter ot write our exercutions.

```python
Count = 0

while Count < 5:
    Count += 1
    print(Count)
```
